```python
import pandas as pd
```

```python
users = pd.read_csv('takehome_users.csv',encoding='latin-1')
```

```python
user_engagement = pd.read_csv('takehome_user_engagement.csv')
```

Defining an **adopted user** as a user who has logged into the product on three separate days in at least one sevenday period, identify which factors predict future user adoption.

```python
user_engagement.time_stamp = pd.to_datetime(user_engagement.time_stamp)
```

```python
user_engagement.time_stamp = user_engagement.time_stamp.apply(lambda x: x.strftime("%Y-%m-%d"))
```

```python
user_engagement.time_stamp = pd.to_datetime(user_engagement.time_stamp)
```

```python
user_engagement.head()
```

```python
user_engagement.set_index(['time_stamp'],inplace=True)
```

```python
user_engagement.head()
```

```python
w = user_engagement.groupby(['user_id'])
```

```python
weekly = w.resample("W").count()
```

```python
users_weekly = weekly[weekly.visited >= 3]
```

```python
users_weekly.sort_index(inplace=True)
```

```python
a = []
for i,j in users_weekly.index.values:
    a.append(i)
```

```python
users_weekly.index.values
```

```python
a = sorted(list(set(a)))
```

```python
a[:10]
```

```python
users.head()
```

```python
users['adopted_user'] = users['object_id'].apply(lambda x: 1 if x in a else 0)
```

```python
users.head(20)
```

```python
users.info()
```

```python
users.last_session_creation_time = users.last_session_creation_time.fillna(users.last_session_creation_time.median())
users.invited_by_user_id = users.invited_by_user_id.fillna(0)
```

```python
users.info()
```

```python
X = users.drop(columns=['creation_time','name','email','adopted_user','last_session_creation_time','object_id'])
y = users['adopted_user']
```

```python
ohe = pd.get_dummies(X['creation_source'])
```

```python
X = X.join(ohe)
X.drop(['creation_source'], axis=1, inplace=True)
```

```python
X.head()
```

```python
y.head()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
X_train
```

```python
cf = RandomForestClassifier(random_state=42)

parameters = {"n_estimators" : [75, 100, 150],
          "max_depth" : [5, 10, 15]}

grid_search = GridSearchCV(estimator=cf,param_grid=parameters,n_jobs=-1,scoring='accuracy',cv=3)

grid_search.fit(X_train,y_train)
y_pred = grid_search.predict(X_test)

accuracy = accuracy_score(y_pred,y_test)

print ('Best score is: ' + str(grid_search.best_score_))
print('Best params : ' + str(grid_search.best_params_))
print("Acuuracy of the tuned model: " + str(accuracy) )
```

```python
cf = RandomForestClassifier()
cf.fit(X_train,y_train)

features = X.columns

for idx, val in enumerate(cf.feature_importances_):
    try:
        print("{:20}{:20f}".format(features[idx], val))
    except IndexError:
            continue
```
